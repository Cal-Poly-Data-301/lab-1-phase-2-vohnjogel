## Question 2

Is there a relationship between the number of siblings a student has, and whether or not they live in SLO?

Make a visualization to address this question, and discuss the result.

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/1DhoCvZVuD0y3P9CuSfyZePizCfJqAvzgk2XiXTOiWYo/edit#gid=970845822'
url_1 = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df_survey = pd.read_csv(url_1)
df_survey['slo'] = df_survey['Where are you living this quarter?'].map({
    'Off Campus, but in SLO': 'yes',
    'Off Campus, outside of SLO': 'no',
    'On-Campus housing': 'yes'
})
joint_slo_sibs = pd.crosstab(df_survey['slo'], df_survey['How many siblings do you have?'])
slo = joint_slo_sibs.sum(axis=1)
cond_slo_sibs = joint_slo_sibs.divide(slo, axis=0)
cond_slo_sibs.T.plot.bar()
```

**There doesn't seem to be much of a relationship between the number of siblings a student has and whether they live in SLO. It's true that most only children are living at home, which makes sense if they don't have to deal with a lot of people at home, but it's not clear to me from this amount of responses that students with more siblings are more likely to live in SLO.**
